```python
class AgeDurationCalculator:
    def __init__(self, age_years):
        self.age_years = age_years

    def calculate_duration(self, unit):
        # Define conversion factors
        conversions = {
            'months': self.age_years * 12,
            'weeks': self.age_years * 52.1775,  # Average weeks in a year
            'days': self.age_years * 365.25,   # Considering leap years
            'hours': self.age_years * 365.25 * 24,
            'minutes': self.age_years * 365.25 * 24 * 60,
            'seconds': self.age_years * 365.25 * 24 * 60 * 60
        }
        return conversions.get(unit.lower(), "Invalid unit")

def main():
    try:
        # Get user input
        age_input = input("What's your age? (in years): ").strip()
        age_years = int(age_input)
        if age_years < 0:
            raise ValueError("Age cannot be negative.")

        unit_input = input("Please choose a time unit (Months, Weeks, Days, Hours, Minutes, Seconds): ").strip().lower()
        unit_map = {
            'm': 'months',
            'w': 'weeks',
            'd': 'days',
            'h': 'hours',
            'min': 'minutes',
            's': 'seconds',
            'months': 'months',
            'weeks': 'weeks',
            'days': 'days',
            'hours': 'hours',
            'minutes': 'minutes',
            'seconds': 'seconds'
        }

        # Resolve the full name of the unit
        unit = unit_map.get(unit_input, None)
        if not unit:
            print("Invalid unit input. Please try again.")
            return

        # Calculate and display result
        calculator = AgeDurationCalculator(age_years)
        result = calculator.calculate_duration(unit)
        print(f"You lived for {result:.2f} {unit.capitalize()}.")

    except ValueError as e:
        print(f"Error: {e}. Please enter a valid age.")

if __name__ == "__main__":
    main()
```
